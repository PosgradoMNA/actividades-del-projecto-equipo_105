###Documentación de parte 3 de código:

Esta sección se encarga de limpiar los datos nulos para asegurar una buena calidad de datos de entrada en el análisis.

Primero se utilizan las funciones estándares de python para hacer una limpieza básica.

Después se utiliza una función hecha a la medida la cuál ocupa como paramétros:


1.   El DataFrame que se quiere limpiar
2.   El porcentaje de los datos más limpios con los cuales quieres hacer el modelo

Esta función se hizo con el propósito que pueda ser utilizada en cualquier caso de negocio, ya que toma como parámetro la propia base de datos y analiza línea por línea qué tantos valores nulos tiene, posteriormente acomoda de "peor a mejor" todas las filas de la base de datos y según el porcentaje que el usuario utilizó como parámetro de "porcentaje de datos más limpios",  la función descarta las peores líneas en la base de datos y devuelve un data frame más limpio.

Como conclusión se va a seguir haciendo este ejercicio utilizando la función hecha a la medida, usando como parámetro el quedarnos con el 99.5% de los datos más limpios, para este caso en particular es lo mismo que eliminar todas las filas con valores nulos.

```python
#Drop the rows that overpass the treshold of null values, if you use the "axis" variable, it indicates it will apply the treshold to the column
ndf = InitialDF.copy()
ndf.dropna(thresh=3,inplace = True)#,axis = 1
ndf
```

```python
##This function outputs a dataframe that has the rows with nulls cleaned depending on the percentile of acceptance defined

def CleanDataFrame(InitialDF,percentile):

  #Get all the headers in the dataframe
  HeadersList = list(InitialDF.columns) #Gets a directory of all headers
  HeadersList

  AllRowsArray = [] #Initialize the array that contains all th rows with nulls

  for i in range(len(HeadersList)): #This loop appends the rows that have a null value into the 'AllRowsArray
    MissingValuesIndex = np.where(InitialDF[HeadersList[i]].isnull())[0]
    AllRowsArray= np.append(AllRowsArray,MissingValuesIndex)
  # print(MissingValuesIndex)

  #print(AllRowsArray)

  from collections import Counter #This library lets me extract unique values as keys and the count of those unique values as values

  IndexesNumber = list(Counter(AllRowsArray).keys()) # equals to list(set(words)), the list() converts it into an array
  CountOfNulls=  list(Counter(AllRowsArray).values()) # counts the elements' frequency, the list() converts it into an array

  ConvertUniqueValues = np.array(IndexesNumber, dtype='i') #converts arrays
  ConvertCountOfNulls = np.array(CountOfNulls, dtype='i')

  UniqueAndCount = np.transpose(np.stack((ConvertUniqueValues, ConvertCountOfNulls)))
  SortedArray = UniqueAndCount[UniqueAndCount[:, 1].argsort()]
  ResultArray = SortedArray[::-1]

  #print(ResultArray)

  EntriesToClean = int((len(InitialDF) * percentile)//1) #Determine how many entries you will drop based on the dataframe size and the chosen percentile

  #print(EntriesToClean)

  if EntriesToClean>len(ResultArray):
    trimmedArray = ResultArray[:,:]
    indexes_to_drop= np.ravel(trimmedArray[:, [0]])
    indexes_to_keep = set(range(InitialDF.shape[0])) - set(indexes_to_drop)
    #print('bigger')
  else:
    trimmedArray = ResultArray[0:-EntriesToClean,0:EntriesToClean]
    indexes_to_drop= np.ravel(trimmedArray[:, [0]])
    indexes_to_keep = set(range(InitialDF.shape[0])) - set(indexes_to_drop)
    #print('smaller')

  SlicedDF = InitialDF.take(list(indexes_to_keep))

  return SlicedDF
```

```python
#Calling CleanDataFrame function

CleanDF= CleanDataFrame(InitialDF,0.99)

print(CleanDF.isna().sum().sum()) #resultado final de filas con datos nulos
CleanDF
```

En este caso se tratarán los datos con la función minmax que normaliza los datos de tal manera que queden en un rango de 0 a 1, sin embargo, antes de normalizar los datos se debe de conocer un poco más lo que representan, por lo tanto primero se procederá a describir un poco las columnas:



###Documentación de parte 6 de código:

Se procederá a normalizar únicamente las columnas numéricas identificadas y se muestra una visualización de los datos en histogramas para ver los resultados de la normalización de los datos numéricos

```python
# copy the data
MinMaxDF = CleanDF.copy()
NumericColumns = ('Credito_otorgado','Edad','Pago_Septiembre','Pago_Agosto','Pago_Julio','Pago_Junio','Pago_Mayo','Pago_Abril','Cantidad_por_pagar_Septiembre',
                  'Cantidad_por_pagar_Septiembre','Cantidad_por_pagar_Agosto','Cantidad_por_pagar_Julio','Cantidad_por_pagar_Junio',
                  'Cantidad_por_pagar_Mayo','Cantidad_por_pagar_Abril','Cantidad_pagada_Septiembre','Cantidad_pagada_Agosto',
                  'Cantidad_pagada_Julio','Cantidad_pagada_Junio','Cantidad_pagada_Mayo','Cantidad_pagada_Abril')
  
# apply normalization techniques
for ListElements in NumericColumns:
    MinMaxDF[ListElements] = (MinMaxDF[ListElements] - MinMaxDF[ListElements].min()) / (MinMaxDF[ListElements].max() - MinMaxDF[ListElements].min())    
  
# view normalized data

Histograms = MinMaxDF.hist(bins=30, figsize=(25, 25))
Histograms
```

```python
#Se vuelven a hacer cálculos del coeficiente de varianza
Coeficiente_Varianza_Columnas2 = (MinMaxDF.std()/(MinMaxDF.mean()))
print('Los coeficientes de varianza de las columnas son:')
Comparacion_coeficientes = pd.concat([Coeficiente_Varianza_Columnas, Coeficiente_Varianza_Columnas2], axis = 1)
Comparacion_coeficientes.rename(columns= {0:'Coeficientes_original',1:'Coeficientes_normalizada'}, inplace = True)
print(Comparacion_coeficientes)
```

Como se puede observar el uso de la técnica de normalización ayudó a algunas columnas a reducir su factor de dispersión, haciendo mucho más útil estas columnas para cualquier análisis.
